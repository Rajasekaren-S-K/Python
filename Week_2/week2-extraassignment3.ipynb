```python
def categorize_age(age):
    if age < 13:
        return "Child"
    elif 13 <= age < 18:
        return "Teenager"
    elif 18 <= age < 30:
        return "Adult"
    elif 30 <= age < 50:
        return "Adult"
    else:
        return "Senior Citizen"
        
ages = int(input("age:"))
print(categorize_age(ages))
```
