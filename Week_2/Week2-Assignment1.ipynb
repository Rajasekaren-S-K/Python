## Body Mass Index

```python
body_mass_value=float(input("Enter the BMI Index :"))
def check_bmi(BMI):
    if (BMI >= 25 and BMI <= 29.9):
        print("Overweight")
    elif(BMI >= 30 and BMI <=34.9):
        print("Class I Obesity")
    elif(BMI >= 35 and BMI <=39.9):
        print("Class II Obesity")    
    elif(BMI > 40):
        print("Class III Obesity")
    elif(BMI < 18.5):
        print("Underweight")
    else:
        print("Normal Weight")

check_bmi(body_mass_value)
```
